# Normalizing Time Series to a Base Value

It's hard to compare stocks' performances when they all start at the different points. We can normalize the starting price to a base value to avoid this.

````python
close.head()
````

````python
# select the price to set as the base
base = close.iloc[0,0]
base
````

````python
# select AAPL col and set it to 100
close.AAPL.div(base).mul(100)
````

````python
# normalize the close df as a whole
start_prices = close.iloc[0]
normalized_close = close.div(start_prices).mul(100)
normalized_close
````

````python
normalized_close.plot()
````

````python
normalized_close.plot(figsize=(15, 8), fontsize=13)
plt.legend(fontsize=13)
plt.show()
````

# Relative change with shift() method
We can use `shift()` method to calculate the relative change within the set window of time (eg. day, week, year)


````python
close.head()
````

````python
# select only AAPL
aapl = close.AAPL.to_frame()
aapl
````

````python
# shift by one index
# first row will be empty
aapl.shift(periods = 1)
````

````python
# Had to drop the column in order to run the next cell again after making some changes
aapl.drop('yesterday',axis=1, inplace=True)
aapl
````

````python
# Adding an extra column with the price lagged by a day
aapl["yesterday"] = aapl.shift(periods=1)
````

````python
aapl
````

````python
aapl["return"] = aapl.AAPL.div(aapl.yesterday).sub(1).mul(100)
````

````python
aapl
````

````python
aapl["change"] = aapl.AAPL.sub(aapl.yesterday)
aapl
````

# Finding abs|rel changes with built-in methods
Pandas provides two built-in methods for getting absolute and relative changes.

````python
aapl["change_by_diff"] = aapl.AAPL.diff(periods=1)
aapl
````

````python
aapl["return_by_pct_change"] = aapl.AAPL.pct_change(periods=1).mul(100)
aapl
````

````python
# get monthly changes with resampling
# resample by the last business day in a month and get pct change since the last record
aapl.AAPL.resample('BM').last().pct_change(periods = 1).mul(100)
````
