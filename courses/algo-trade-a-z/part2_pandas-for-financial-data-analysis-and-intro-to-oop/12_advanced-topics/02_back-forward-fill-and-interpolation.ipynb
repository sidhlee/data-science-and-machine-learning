# Back|Forward-filling missing values

```python
import pandas as pd
```

```python
stocks = pd.read_csv('stocks.csv', header=[0, 1], parse_dates=[0], index_col=[0])
```

```python
close = stocks.Close
close
# this only incudes business days
```

```python
close["Day"] = close.index.day_name()
close["Quarter"] = close.index.quarter
close
```

```python
# Include non-business days
datetime_index = pd.date_range(start='2009-12-31', end='2019-02-05')
datetime_index
```

```python
# reindex using date range
close = close.reindex(datetime_index)
close
```

```python
# Backfill Day value
close.Day = close.index.day_name()
close.head()
```

```python
# Backfill quarter value
close.Quarter = close.index.quarter
close.head()
```

```python
# forward-fill missing prices
# ALWAYS USE FFILL for stock prices!
close.fillna(method='ffill').head()
```

```python
close.head()
```

```python
close.dropna(inplace=True)
```

```python
# We could also fill the missing value when re-indexing
# (but then we would still need to override day and quarter)
close.reindex(datetime_index, method='ffill')
```

```python
temp = pd.read_csv('temp.csv', index_col="datetime", parse_dates=["datetime"])
temp.head()
```

```python
# Resample the temp data in 30min
temp = temp.resample("30 Min").mean()
temp.head()
```

```python
# Another option for filling missing value - interpolation
# we can derive the missing value from the neighboring values (default - linear interpolation)
temp.interpolate()
```
