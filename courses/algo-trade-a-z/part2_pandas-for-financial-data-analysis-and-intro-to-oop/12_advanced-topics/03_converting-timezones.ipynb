# Timezones and conversions

```python
import pandas as pd
```

```python
ge = pd.read_csv('GE_prices.csv', parse_dates=["date"], index_col="date")
ge
```

```python
# we can see that the data starts at 10am and ends at 4pm -> this is EST, not UTC
ge.loc["2019-07-29"]
```

```python
print(ge.index.tz) # no timezone info
```

```python
# make the df timezone-aware using tz_localize
ge = ge.tz_localize('America/New_York')
```

```python
ge.head()
```

```python
ge.index.tz
```

```python
ge.tz_convert('UTC').head()
```

```python
import pytz 
```

```python
pytz.common_timezones
```

```python
import re
[tz for tz in pytz.all_timezones if re.search(r'seoul', tz, re.IGNORECASE)]
```
